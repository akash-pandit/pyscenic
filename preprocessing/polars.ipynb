```python
import polars as pl

from pathlib import Path

PROJROOT = Path.cwd().parent
RAW_DATA = PROJROOT / "data/raw"
PROCESSED_DATA = PROJROOT / "data/preprocessed"
DATABASES = PROJROOT / "databases"

RESULTS = PROJROOT / "results"
```

```python
query = (
    pl.scan_csv(RESULTS / "*/grn/adj.tsv", separator='\t')
    .select(['TF', 'target', 'importance'])
    .group_by(['TF', 'target'])
    .agg( pl.col('importance') )
    .with_columns(
        importance=pl.col('importance').list.mean()
    )
)

avg_adj = query.collect()
```

```python
avg_adj.sort('importance')
```

```python
avg_adj.write_csv("adj.tsv", separator='\t')
```
